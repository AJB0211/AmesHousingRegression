```python
def checkCols(x,colnames):
    @np.vectorize
    def compare(x,y):
        try: 
            return np.equal(x,y)
        except:
            return x == y
    return np.all([compare(x[i+"_x"],x[i+"_y"]) for i in colnames])
```
